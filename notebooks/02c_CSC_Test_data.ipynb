XYZ to LAB

```python
LAB1 = XYZ1.to_LAB()
LAB2 = XYZ2.to_LAB()
```

```python
AEab = LAB1.delta_e_ab(LAB2)
AEab
```

```python
AEuv = LUV2.delta_e_uv(LUV1)
AEuv
```
