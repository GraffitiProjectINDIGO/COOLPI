## CIE Observer

```python
from coolpi.colour.cie_colour_spectral import Observer
```

### To create an instance:

The *Observer* class receives a valid `str` or `int` type for the specification of the 2º or 10º degree CIE standard observer:

2º standard observer (CIE 1931)

```python
obs_2 = Observer(2) # 2 or "2"
```

```python
type(obs_2)
```

10º standard observer (CIE 1964)

```python
obs_10 = Observer("10") # 10 or "10"
```

In case the observer is not specified, the default value is 2 (so an instance of the 2º standard colorimetric observer is created):

```python
obs = Observer()
```

```python
print(obs)
```

It is only allowed to create instances for CIE standard observers 2º and 10º degree. Otherwise, a `CIEObserverError` is raised:

```python
Obs = Observer(3)
```

### Attributes

```python
print(obs_2.type)
print(obs_2.subtype)
print(obs_2.observer)
```

### Method

The `__str__` method returns the string representation of the object. This method is called when the print() or str() 
function is invoked on an object.

```python
print(obs)
```

## SComponents

```python
from coolpi.colour.cie_colour_spectral import SComponents
```

### To create an instance:

No parameters are required to instantiate the class:

```python
sc = SComponents()
```

### Attributes

```python
print(sc.type)
print(sc.subtype)
print(sc.nm_range)
print(sc.nm_interval)
```

Each of the $S_0(\lambda)$, $S_1(\lambda)$ and $S_2(\lambda)$ components can be accessed directly as follows:

```python
S0 = sc.S0
S1 = sc.S1
S2 = sc.S2

print(S0.name_id, S0.nm_range, S0.nm_interval)
print(S1.name_id, S1.nm_range, S1.nm_interval)
print(S2.name_id, S2.nm_range, S2.nm_interval)
```

### Methods:

The `__str__` method returns the string representation of the object. This method is called when the print() or str() 
function is invoked on an object.

```python
print(sc)
```

`To get the spectral *SComponents*`:

An alternative way to access the components is by using the following class method:

```python
S0, S1, S2 = sc.get_S_components()

print(S0.name_id, S0.nm_range, S0.nm_interval)
print(S1.name_id, S1.nm_range, S1.nm_interval)
print(S2.name_id, S2.nm_range, S2.nm_interval)
```

Attributes:

```python
print(S0.name_id)
print(S0.nm_range)
print(S0.nm_interval)
print(S0.lambda_values)
```

For some calculations, the spectral values of the components can be accessed directly using the method `get_S_components_lambda_values`.

```python
S0_lambda, S1_lamdda, S2_lambda = sc.get_S_components_lambda_values()
```

`To compute the SPD of an illuminant from a CCT in º Kelvin`:

```python
nm_range, nm_interval, spd = sc.CCT_to_SPD(6500)
```

```python
print(nm_range)
print(nm_interval)
print(spd)
```

Plot illuminant

```python
from coolpi.auxiliary.plot import plot_illuminant

illuminant = {"6500ºK":(nm_range, nm_interval, spd)}
plot_illuminant(illuminant, show_figure = True)
```

### Plot

To plot the $S_0(\lambda)$, $S_1(\lambda)$ and $S_2(\lambda)$ components:

```python
sc.plot()
```

To save the figure:

It is possible to save the figure by specifying `save_figure` = `True`, and entering the `output_path` as follows:

```python
path_figure = os.path.join(*["res", "plot", "s_components.png"]) # your path here
sc.plot(show_figure = False, save_figure = True, output_path = path_figure)
```

## CIE CMF

```python
from coolpi.colour.cie_colour_spectral import CMF
```

### To create an instance:

The *CMF* class receives a valid `str` or `int` type for the specification of the  2º or 10º degree CIE standard observer:

```python
cmf_2 = CMF(2)
```

In case the observer is not specified, the default value is 2 (so an instance of the *CMF* for the CIE 1931 
2º standard colorimetric observer is created):

```python
cmf_2 = CMF()
print(cmf_2)
```

Also, the *CMF* class can be instantiated using an *Observer* instance as input parameter:

```python
obs_10 = Observer(10)
cmf_10 = CMF(obs_10)
```

### Attributes

```python
print(cmf_2.type)
print(cmf_2.subtype)
print(cmf_2.nm_range)
print(cmf_2.nm_interval)
print(cmf_2.observer)
```

```python
print(cmf_10.type)
print(cmf_10.subtype)
print(cmf_10.nm_range)
print(cmf_10.nm_interval)
print(cmf_10.observer)
```

Each of the CMFs can be accessed directly as follows:

```python
x_cmf = cmf_2.x_cmf
y_cmf = cmf_2.y_cmf
z_cmf = cmf_2.z_cmf

print(x_cmf.name_id, x_cmf.nm_range, x_cmf.nm_interval)
print(y_cmf.name_id, y_cmf.nm_range, y_cmf.nm_interval)
print(y_cmf.name_id, y_cmf.nm_range, y_cmf.nm_interval)
```

### Methods

The `__str__` method returns the string representation of the object. This method is called when the print() or str() 
function is invoked on an object.

```python
print(cmf_2)
```

`To get the colour-matching-functions`:

An alternative way to access the color-matching-functions is by using the class method:

```python
x_cmf, y_cmf, z_cmf = cmf_2.get_colour_matching_functions()

print(x_cmf.name_id, x_cmf.nm_range, x_cmf.nm_interval)
print(y_cmf.name_id, y_cmf.nm_range, y_cmf.nm_interval)
print(y_cmf.name_id, y_cmf.nm_range, y_cmf.nm_interval)
```

### Plot

To plot the $\bar{x}(\lambda)$, $\bar{y}(\lambda)$ and $\bar{z}(\lambda)$ colour-matching-functions:

```python
cmf_2.plot()
```

```python
cmf_10.plot()
```

```python
path_figure = os.path.join(*["res", "plot", "cmf_10_plot.png"]) # your path here
cmf_10.plot(show_figure = False, save_figure = True, output_path = path_figure)
```

`Extrapolation`

```python
cmf_2.cmf_extrapolate([340,830], 1, method="Spline")
cmf_2.plot()
```

```python
path_figure = os.path.join(*["res", "plot"], "cmf_2_extrapolate_plot.png") # your path here
cmf_2.plot(show_figure = False, save_figure = True, output_path = path_figure)
```
